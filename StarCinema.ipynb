```python
#Homework1
class Movie:
    def __init__(self, name, genre, duration):
        self.name = name
        self.genre=genre
        self.duration=duration
    def movieInfo(self):
        print('Movie Name:',self.name)
        print('Movie Genre:',self.genre)
        print('Movie Duration:',self.duration,'minutes.')
    @classmethod
    def createMovie_fromString(cls,string1):
        name,genre,duration=string1.split('-')
        obj=cls(name,genre,duration)
        return obj

class StarCinema:
    all_branch_info={}
    def __init__(self,branch_name):
        self.branch_name=branch_name
        self.movie_list=[]
        print(f'Welcome to the {self.branch_name} branch of StarCinema!')
    def addMovies(self,*movie_objects):
        for i in movie_objects:
            if i not in self.movie_list:
                self.movie_list.append(i)
                StarCinema.all_branch_info[self.branch_name]=self.movie_list
                print(f"{i.name} added to {self.branch_name} branch")
            else:
                print('The movie is already added in this branch.')
    def removeMovie(self,obj1):
        for i,j in StarCinema.all_branch_info.items():
            if self.branch_name in i:
                j.remove(obj1)
    @classmethod
    def check(cls,movie_name):
        switch=False
        for i,j in StarCinema.all_branch_info.items():
            for x in j:
                if movie_name==x.name:
                    switch=True
            if switch:
                print(f'{movie_name} is being streamed in {i} branch')
            else:
                print(f'{movie_name} is not being streamed in any branch.')
    @classmethod
    def showAllBranchInfo(cls):
        for i,j in StarCinema.all_branch_info.items():
            print('Branch Name:',i)
            for index, movies in enumerate(j,start=1):
                print('Movie No:',index)
                print('Movie Name:',movies.name)
                print('Movie Genre:',movies.genre)
                print('Movie Duration',movies.duration,'minutes')
                print('-------------------------')
            print('**************************')

movie1 = Movie('Godfather', 'Crime Drama', 172)
movie2 = Movie('Godfather Part-II', 'Crime Drama', 184)
movie3 = Movie('Godfather Part-III', 'Crime Drama', 168)
print('1==========================================')
movie3.movieInfo()
print('2==========================================')
movie4 = Movie.createMovie_fromString('Goodfellas-Biographical-174')
print('3==========================================')
movie4.movieInfo()
print('4=========================================')
branch1 = StarCinema('Bashundhora')
print('5==========================================')
branch1.addMovies(movie1, movie2, movie4)
print('6==========================================')
branch1.addMovies(movie1, movie3)
print('7==========================================')
StarCinema.showAllBranchInfo()
print('8==========================================')
branch2 = StarCinema('Shimanto')
print('9==========================================')
branch2.addMovies(movie1, movie2, movie3)
print('10==========================================')
StarCinema.showAllBranchInfo()
print('11==========================================')
StarCinema.check('Godfather')
print('12=========================================')
StarCinema.check('Barbie')
print('13=========================================')
branch1.removeMovie(movie4)
StarCinema.showAllBranchInfo()
print('14=========================================')
branch2.removeMovie(movie1)
StarCinema.showAllBranchInfo()

```
