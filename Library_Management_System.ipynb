```python
class Book:
    def __init__(self, title, author, genre):
        self.__title = title
        self.__author = author
        self.__genre = genre
        self.__available = True
        self.__borrower = None

    def display_info(self):
        print('Title:',self.__title)
        print('Author:',self.__author)
        print('Genre:',self.__genre)
        print('Availability:',self.__available)
        print('---------------')

    def set_availability(self, available):
        self.__available = available

    def get_availability(self):
        return self.__available

    def get_borrower(self):
        return self.__borrower

    def set_borrower(self, member):
        self.__borrower = member


class LibraryMember:
    def __init__(self, id, name):
        self.__id = id
        self.__name = name
        self.__borrowed_books = []

    def borrow_book(self, book):
        if book.get_availability():
            self.__borrowed_books.append(book)
            book.set_availability(False)
            book.set_borrower(self)


    def return_book(self, book):
        if book in self.__borrowed_books:
            self.__borrowed_books.remove(book)
            book.set_availability(True)
            book.set_borrower(None)


    def display_borrowed_books(self):
        print(f"Borrowed books by {self.__name}:")
        for book in self.__borrowed_books:
            book.display_info()


class Library:
    def __init__(self):
        self.__books_available = []
        self.__library_members = []

    def add_book(self, book):
        self.__books_available.append(book)

    def add_library_member(self, member):
        self.__library_members.append(member)

    def display_book_list(self):
        print("All the available Books in library:")
        for book in self.__books_available:
            book.display_info()

    def display_library_members(self):
        print("Library Members:")
        for member in self.__library_members:
            print(f"{member._LibraryMember__name} ({member._LibraryMember__id})")



book1 = Book("The Alchemist", "Paulo Coelho", "Fiction")
book2 = Book("Atomic Habits","James Clear","Non-Fiction")
book3 = Book("The Elephant in the Brain", "Kevin Simler", "Non-Fiction")

member1 = LibraryMember("LM01", "Rohim")
member2 = LibraryMember("LM02", "Korim")

library = Library()

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.add_library_member(member1)
library.add_library_member(member2)

member1.borrow_book(book1)
member1.borrow_book(book2)
member2.borrow_book(book3)

library.display_book_list()
print("1======================================")

member1.display_borrowed_books()
print("2======================================")
member2.display_borrowed_books()
print("3======================================")

member1.return_book(book2)
print("4======================================")

member1.display_borrowed_books()
library.display_book_list()
```

```python
#ClassWork2
class Spaceship:
    def __init__(self,ship_name,capacity):
        self.ship_name=ship_name
        self.capacity=capacity
        self.loaded_cargo_name=[]
        self.loaded_cargo_weight=0
    def load_cargo(self,obj):
        if self.loaded_cargo_weight+obj.get_weight() <= self.capacity:
            self.loaded_cargo_name.append(obj.get_cargo_name())
            self.loaded_cargo_weight+=obj.get_weight()
        else:
            print(f'Warning: Unable to load Neutronium inside {self.ship_name}. Exceeds capacity by {self.loaded_cargo_weight}.')
    def display_details(self):
        print('Spaceship Name:',self.ship_name)
        print('Capacity:',self.capacity)
        print('Current Cargo Weight:',self.loaded_cargo_weight)
        print('Cargo:',self.loaded_cargo_name)
class Cargo:
    def __init__(self,cargo_name,weight):
        self.__cargo_name=cargo_name
        self.__weight=weight
    def get_weight(self):
        return self.__weight
    def get_cargo_name(self):
        return self.__cargo_name

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#Classwork3
class Song:
    def __init__(self, title, duration, artist="Unknown Artist"):
        self.__title = title
        self.__artist = artist
        self.__duration = duration

    def set_title(self, title):
        self.__title = title

    def set_artist(self, artist):
        self.__artist = artist

    def set_duration(self, duration):
        self.__duration = duration

    def get_title(self):
        return self.__title

    def get_artist(self):
        return self.__artist

    def get_duration(self):
        return self.__duration

    def info(self):
        minutes = self.__duration // 60
        seconds = self.__duration % 60
        return f"{self.__title} by {self.__artist} ({minutes}:{seconds:02})"


class Playlist:
    def __init__(self, name, *songs):
        self.__name = name
        self.__song_list = list(songs)
        self.__now_id = -1

    def set_name(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def play_next(self):
        if self.__now_id == len(self.__song_list) - 1:
            print("Reached the end of the playlist, starting from the first song.")
            self.__now_id = 0
        else:
            self.__now_id += 1
        print(f"Now playing: {self.__song_list[self.__now_id].info()}")

    def play_prev(self):
        if self.__now_id == 0:
            print("Cannot go to previous from the first song.")
        else:
            self.__now_id -= 1
        print(f"Now playing: {self.__song_list[self.__now_id].info()}")

    def add_songs(self, *songs):
        self.__song_list.extend(songs)

    def remove_song(self, song_obj):
        if song_obj in self.__song_list:
            self.__song_list.remove(song_obj)

    def show_playlist(self):
        print(f"Playlist: {self.__name}")
        print(f"Total Songs: {len(self.__song_list)}")
        for i in range(len(self.__song_list)):
            print(i,'.',self.__song_list[i].info())


# Driver Code
s1 = Song("The Flower Duet", 86, "Leo Delibes")
print(s1.info())

print('===============')

s2 = Song("Ride of the Valkyries", 137, "Wagner")
s3 = Song("Russian Dance - Nutcracker", 71)

p1 = Playlist("Classical Orchestra", s1, s2, s3)
p1.show_playlist()
print('===============')

p1.play_next()
print('===============')
p1.play_prev()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
```
