```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group
        self.next = None
        self.prev = None


class WRM:
    def __init__(self):
        self.head = None

    def register_patient(self, id, name, age, blood_group):
        new_patient = Patient(id, name, age, blood_group)

        if self.head is None:
            self.head = new_patient
            self.head.next = self.head
            self.head.prev = self.head
        else:
            last_patient = self.head.prev
            last_patient.next = new_patient
            new_patient.prev = last_patient
            new_patient.next = self.head
            self.head.prev = new_patient

    def serve_patient(self):
        if self.head is None:
            print("No patients in the waiting room.")
        else:
            served_patient = self.head
            if self.head.next == self.head:
                self.head = None
            else:
                self.head = self.head.next
                self.head.prev = served_patient.prev
                served_patient.prev.next = self.head

            print("Serving patient:", served_patient.name)

    def cancel_all(self):
        self.head = None
        print("All appointments canceled.")

    def can_doctor_go_home(self):
        if self.head is None:
            return True
        else:
            return False

    def show_all_patients(self):
        if self.head is None:
            print("No patients in the waiting room.")
        else:
            current_patient = self.head
            while True:
                print("Patient:", current_patient.name)
                current_patient = current_patient.next
                if current_patient == self.head:
                    break

    def reverse_the_line(self):
        if self.head is None:
            print("No patients in the waiting room.")
        else:
            current_patient = self.head.prev
            while True:
                print("Patient:", current_patient.name)
                current_patient = current_patient.prev
                if current_patient == self.head.prev:
                    break


class Tester:
    def __init__(self):
        self.waiting_room = WRM()

    def menu(self):
        print("Options:")
        print("a. Add Patient")
        print("b. Serve Patient")
        print("c. Show All Patients")
        print("d. Can Doctor Go Home?")
        print("e. Cancel All Appointments")
        print("f. Reverse The Line")
        print("q. Quit")

    def add_patient(self):
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")

        self.waiting_room.register_patient(id, name, age, blood_group)
        print("Patient added successfully.")

    def serve_patient(self):
        self.waiting_room.serve_patient()

    def show_all_patients(self):
        self.waiting_room.show_all_patients()

    def can_doctor_go_home(self):
        if self.waiting_room.can_doctor_go_home():
            print("Yes, doctor can go home.")
        else:
            print("No, doctor cannot go home.")

    def cancel_all_appointments(self):
        self.waiting_room.cancel_all()

    def reverse_the_line(self):
        self.waiting_room.reverse_the_line()

    def run(self):
        while True:
            self.menu()
            option = input("Enter an option: ")

            if option == "a":
                self.add_patient()
            elif option == "b":
                self.serve_patient()
            elif option == "c":
                self.show_all_patients()
            elif option == "d":
                self.can_doctor_go_home()
            elif option == "e":
                self.cancel_all_appointments()
            elif option == "f":
                self.reverse_the_line()
            elif option == "q":
                break
            else:
                print("Invalid option. Please try again.")


tester = Tester()
tester.run()
```
