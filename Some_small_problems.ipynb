```python
#Task01
class Fruit:
    def __init__(self, formalin=False, name=''):
        self.__formalin = formalin
        self.name = name
    def getName(self):
        return self.name
    def hasFormalin(self):
        return self.__formalin
class testFruit:
    def test(self, f):
        print('----Printing Detail----')
        if f.hasFormalin():
            print('Do not eat the',f.getName(),'.')
            print(f)
        else:
            print('Eat the',f.getName(),'.')
            print(f)
class Mango(Fruit):
    def __init__(self, formalin=False, name=''):
        super().__init__(formalin=True, name='Mango')
    def __str__(self):
        if self.hasFormalin():
            return f'{self.name}s are bad for you'
        else:
            return f'{self.name}s are good for you'
class Jackfruit(Fruit):
    def __init__(self, formalin=False, name=''):
       super().__init__(formalin=False, name='Jackfruit')
    def __str__(self):
        if self.hasFormalin():
            return f'{self.name}s are bad for you'
        else:
            return f'{self.name}s are good for you'

m = Mango()
j = Jackfruit()
t1 = testFruit()
t1.test(m)
t1.test(j)
```

```python
#Task2
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self, marks, time, *subjects):
        super().__init__(marks)
        self.time = time
        self.subjects = subjects

    def examSyllabus(self):
        syllabus = super().examSyllabus()
        if self.subjects:
            syllabus += " , " + " , ".join(self.subjects[:-1])
            if len(self.subjects) > 1:
                syllabus += " ,"
            syllabus += " " + self.subjects[-1]
        return syllabus

    def examParts(self):
        parts = super().examParts()
        for i in range(len(self.subjects)):
            parts += f"\nPart {i + 3} - {self.subjects[i]}"
        return parts
    def __str__(self):
        return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {len(self.subjects) + 2}"
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#Task3
class PokemonBasic:

  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type

  def get_type(self):
    return 'Main type: ' + self.type

  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'

  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness

class PokemonExtra(PokemonBasic):
    def __init__(self, name, hp, weakness, type, secondary_type=None, moves=None):
        super().__init__(name, hp, weakness, type)
        self.secondary_type = secondary_type
        self.moves = moves

    def get_type(self):
        main_type = super().get_type()
        if self.secondary_type:
            return main_type + ', Secondary type: ' + self.secondary_type
        return main_type

    def get_move(self):
        basic_move = super().get_move()
        moves_str = ""

        if self.moves:
            for i in range(len(self.moves)):
                if i > 0:
                    moves_str += ", "
                moves_str += self.moves[i]
            moves_str = "\nOther move: " + moves_str

        return basic_move + moves_str

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())

print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#Task4
class Cakes:
    regular_cake_price = 1200
    order_list = {}
    feedbacks = {}
    def __init__(self, flavor, weight):
        self.flavor = flavor
        self.weight = weight
        self.sweetness = "Moderate sugar"
        self.cream_type = "Whipped Cream"
        self.customization = "Not customized"
        self.price = self.calculate_price()

        if flavor in Cakes.order_list:
            Cakes.order_list[flavor] += weight
        else:
            Cakes.order_list[flavor] = weight

    def calculate_price(self):
        return (self.weight / 1000) * Cakes.regular_cake_price

    def add_customization(self, sweetness="Zero", cream_type="Butter Cream"):
        self.sweetness = sweetness
        self.cream_type = cream_type
        self.customization = "Customized"

    def cake_details(self):
      cake_type = "Cake"
      if type(self) == Cheese_Cakes:
          cake_type = "Cheese Cake"
      price_str = str(int(self.price))
      if self.price == int(self.price):
          price_str = str(int(self.price))
      else:
          price_str = str(self.price)
      print("Flavor:", self.flavor, cake_type + ",", "Weight:", self.weight, "gm")
      print("Sweetness:", self.sweetness + ",", "Cream Type:", self.cream_type)
      if type(self) == Cheese_Cakes:
          print("Cake Type:", self.cake_type, end=", ")
      print("Price:", price_str, "Taka")

      order_details = f"{self.flavor}cake {self.weight}gm:1"
      if order_details in Cakes.order_list:
          return Cakes.order_list[order_details]



    @classmethod
    def give_feedbacks(cls, flavor, feedback):
        if flavor in cls.feedbacks:
            cls.feedbacks[flavor].append(feedback)
        else:
            cls.feedbacks[flavor] = [feedback]
        print("Thanks for your valuable feedback!")
        if issubclass(cls, Cheese_Cakes):
            print("You will get 10% discount for your next purchase!")

    @classmethod
    def show_feedbacks(cls):
        for flavor, feedbacks in cls.feedbacks.items():
            print(f"{flavor}: {feedbacks}")


class Cheese_Cakes(Cakes):
    cheese_cake_price = 1500

    def __init__(self, flavor, weight, cake_type="Baked"):
        super().__init__(flavor, weight)
        self.cream_type = "Cream Cheese"
        self.cake_type = cake_type
        self.price = self.calculate_cheese_cake_price()

    def calculate_cheese_cake_price(self):
        return (self.weight / 1000) * Cheese_Cakes.cheese_cake_price

    def add_customization(self):
        print("Sorry! No customization available for cheese cakes.")

order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()

```
