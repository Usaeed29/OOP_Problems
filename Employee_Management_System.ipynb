```python

class Employee:
    employee_count = 0

    def __init__(self, name, joining_date, work_experience, weekly_work_hour):
        self.name = name
        self.joining_date = joining_date
        self.work_experience = work_experience
        self.weekly_work_hour = weekly_work_hour
        Employee.employee_count += 1

    @classmethod
    def showDetails(cls):
        print(f"Company workforce:\nTotal Employee/s: {cls.employee_count}")

class Programmer(Employee):
    total_programmer_count = 0

    def __init__(self, name, joining_date, work_experience, weekly_work_hour=40):
        super().__init__(name, joining_date, work_experience, weekly_work_hour)
        self.designation = ""
        self.salary = 0
        Programmer.total_programmer_count += 1

    def showProgrammerDetails(self):
        print(f"Programmer Employee:\nName: {self.name}\nID: {self.createProgrammerID()}\nJoining Date: {self.joining_date}\nDesignation: {self.designation}\nSalary: BDT {self.salary}")

    def calculateSalary(self):
        base_salaries = {"Junior Software Engineer": 30000,"Software Engineer": 45000,"Senior Software Engineer": 70000,"Technical Lead": 120000}
        self.designation = self.calculateDesignation()
        self.salary = base_salaries[self.designation]
        years = (2023 - int(self.joining_date.split("-")[0]))
        for i in range(years):
            self.salary += self.salary * 0.15

    def calculateDesignation(self):
        if self.work_experience < 2:
            return "Junior Software Engineer"
        elif self.work_experience < 5:
            return "Software Engineer"
        elif self.work_experience < 8:
            return "Senior Software Engineer"
        else:
            return "Technical Lead"
    def createProgrammerID(self):
        date_parts = self.joining_date.split("-")
        year_part = date_parts[0][2:]
        id_format = f"P-{year_part}{date_parts[1]}{date_parts[2]}-{Employee.employee_count}"
        return id_format

    def calculateOvertime(self):
        overtime_hours = max(self.weekly_work_hour - 40, 0)
        overtime_amount = overtime_hours * 500 *4
        self.salary += overtime_amount
        print(f"{self.name} will get BDT {overtime_amount} overtime.")


class HR(Employee):
    total_hr_count = 0

    def __init__(self, name, joining_date, work_experience, weekly_work_hour=40):
        super().__init__(name, joining_date, work_experience, weekly_work_hour)
        self.hr_id = self.createHREmployeeID()
        HR.total_hr_count += 1

    def showHREmployeeDeatails(self):
        print(f"HR Employee:\nName:, {self.name}\nID: {self.hr_id}\nJoining Date:{self.joining_date}")
    def createHREmployeeID(self):
        date_parts = self.joining_date.split("-")
        year_part = date_parts[0][2:]
        id_format = f"HR-{year_part}{date_parts[1]}{date_parts[2]}-{Employee.employee_count}"
        return id_format


class InternProgrammer(Programmer):
    intern_count = 0

    def __init__(self, name, joining_date, intern_type="Unpaid"):
        super().__init__(name, joining_date, work_experience=0)
        InternProgrammer.intern_count += 1
        self.temp_id = f"Temp_{InternProgrammer.intern_count}"
        self.intern_type = intern_type
        self.status = "Eligible for promotion"

    def showInternDetails(self):
        print(f"Intern (Programmer):\nName: {self.name}\nID: {self.temp_id}\nJoining Date: {self.joining_date}\nType: {self.intern_type}\nStatus: {self.status}")
    def promoteToProgrammer(self):
        if self.status == "Eligible for promotion":
            print(f"The intern {self.name} is promoted!")
            programmer = Programmer(self.name, "2023-08-12", work_experience=0)
            return programmer
        else:
            print(f"The intern {self.name} can not be promoted.")

Employee.showDetails()
print("=========1=========")
richard = Programmer("Richard Hendricks", "2021-06-08", 4, 48)
richard.calculateSalary()
print("=========2=========")
richard.showProgrammerDetails()
print("=========3=========")
richard.calculateOvertime()
print("=========4=========")
richard.showProgrammerDetails()
print("=========5=========")
monica = HR("Monica Hall", "2022-07-06", 2, 40)
print("=========6=========")
monica.showHREmployeeDeatails()
print("=========7=========")
Employee.showDetails()
print("=========8=========")
gilfoyle = Programmer("Bertram Gilfoyle", "2020-03-02", 6, 35)
gilfoyle.calculateSalary()
print("=========9=========")
gilfoyle.calculateOvertime()
print("=========10=========")
gilfoyle.showProgrammerDetails()
print("=========11=========")
gavin = Programmer("Gavin Belson", "2016-12-20", 9)
gavin.calculateSalary()
gavin.calculateOvertime()
gavin.showProgrammerDetails()
print("=========12=========")
yang = InternProgrammer("Jian Yang", "2023-01-01")
yang.showInternDetails()
print("=========13=========")
jared = InternProgrammer("Jared Dunn", "2023-06-05", "Paid")
jared.showInternDetails()
print("=========14=========")
jared = jared.promoteToProgrammer()
print("=========15=========")
Employee.showDetails()
print("=========16=========")
yang = yang.promoteToProgrammer()
yang.calculateSalary()
yang.showProgrammerDetails()
print("=========17=========")
Employee.showDetails()
```
